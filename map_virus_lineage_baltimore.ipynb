```python
import pandas as pd
```

```python
df_vir = pd.read_excel("Acronyms.xlsx").dropna(axis=1,how="all") # sheet 0 o 1 ?
display(df_vir)
df_vir = df_vir[['Family', 'Genus',  'Genome Composition']].drop_duplicates()
print(df_vir.nunique())
print(df_vir.shape[0])
print(df_vir["Genome Composition"].value_counts())
## drop pos/neg sense resolution. Could add RT spec. as third baltime family
df_vir["Genome Composition"] = df_vir["Genome Composition"].str.replace("(+/-)","",regex=False).str.replace("\(+/-\)","")
df_vir["Baltimore"] = df_vir["Genome Composition"].str.replace(r'\(\+/-\)|\(\-\)|\(\+\)', '', regex=True)
print(df_vir["Baltimore"].value_counts())
df_vir = df_vir.drop_duplicates(["Family","Baltimore"]) ## drops some rows! 
display(df_vir)
```

```python
## could work on meta direct, or on pred results
df_meta = pd.read_csv("SWP_humVir_meta.csv.gz") ## all # ORIG
# df_meta = pd.read_csv("SWP_humVir_meta_v2.csv.gz") ## all
print(df_meta.shape[0]," # rows in all (meta)data")
## keep only viruses, +- host
df = df_meta.loc[df_meta["Taxonomic lineage"].str.contains("vir",case=False,na=False)].copy()
df = df.filter([ 'Protein names', 'Organism', #'Keywords',  # 'UR50_Cluster ID', 
       'Virus hosts', 'Taxonomic lineage', 'Sequence'],axis=1).copy()
print(df.shape[0],"# vir rows")
df
```

```python
# Initialize empty lists to store extracted family and genus values
family_list = []
genus_list = []

# Loop through each row in the 'Taxonomic lineage' column
for lineage in df["Taxonomic lineage"]:
    # Initialize variables to hold the family and genus for each row
    family = None
    genus = None
    
    # Split the text by commas and loop through each component
    for component in lineage.split(','):
        # Strip leading and trailing whitespaces
        component = component.strip()
        
        # Check for the presence of 'family' and 'genus' and extract their values
        if '(family)' in component:
            family = component.replace('(family)', '').strip()
        elif '(genus)' in component:
            genus = component.replace('(genus)', '').strip()
    
    # Append the extracted values to the respective lists
    family_list.append(family)
    genus_list.append(genus)

# Add new columns to the DataFrame
df['Family'] = family_list
df['Genus'] = genus_list
```

#### Manually fox some missing families
Some viruses missing from the viral zone downloaded file. (Despite appearing in site). 
 * I manually update them
* Sedoreoviridae =  dsRNA viruses ; https://viralzone.expasy.org/104 , etc....
*  Kolmioviridae - Circular complementary ssRNA viruses ; https://viralzone.expasy.org/9776 ; Circular, ssRNA(-)

* Nora virus - I assume is actualy Norovirus  (ssRNA positive-strand viruses)

```python
df.loc[df["Family"]=="Sedoreoviridae",['Baltimore',"Genome Composition"]] = "dsRNA"
df.loc[df["Family"]=="Spinareoviridae",['Baltimore',"Genome Composition"]] = "dsRNA"
df.loc[df["Family"]=="Tobaniviridae",'Baltimore'] = "ssRNA"
df.loc[df["Family"]=="Tobaniviridae","Genome Composition"] = "ssRNA(+)"

df.loc[df["Family"]=="Polydnaviriformidae",['Baltimore',"Genome Composition"]] = "dsDNA"

df.loc[df["Family"]=="Matonaviridae",'Baltimore'] = "ssRNA"
df.loc[df["Family"]=="Matonaviridae","Genome Composition"] = "ssRNA(+)"

df.loc[df["Family"]=="Kolmioviridae",'Baltimore'] = "ssRNA"
df.loc[df["Family"]=="Kolmioviridae","Genome Composition"] = "ssRNA(-)"

### Nora virus - is presumably Norovirus
df.loc[df["Organism"]=="Nora virus",'Family'] = "Caliciviridae"
df.loc[df["Organism"]=="Nora virus",'Genus'] = "Norovirus"
df.loc[df["Organism"]=="Nora virus",'Baltimore'] = "ssRNA"
df.loc[df["Organism"]=="Nora virus","Genome Composition"] = "ssRNA(+)"


print(f"{df['Baltimore'].isna().sum()-s1}# rows missing family match")
```
